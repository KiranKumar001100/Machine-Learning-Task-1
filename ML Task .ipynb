```python
# Describing Data set 
df1.describe(include = 'all')
```

# Q1. Mean Function #

```python
df1.price_per_sqft.mean()
```

# Q2. Percentile Method #

```python
Q1 = df1['price_per_sqft'].quantile(0.25)
Q2 = df1['price_per_sqft'].quantile(0.50)
Q3 = df1['price_per_sqft'].quantile(0.75)
Q1, Q2, Q3
```

# Q3. IQR #

```python
IQR = Q3 - Q1
IQR
```

```python
# Find Outliers
lower_whisker = Q1-1.5*IQR
lower_whisker
```

```python
upper_whisker = Q3+1.5*IQR
upper_whisker
```

```python
outliers = df1[(df1.price_per_sqft < -411.5) | (df1.price_per_sqft > 12184.5)]
outliers
```

```python
non_outliers = df1[(df1.price_per_sqft > -411.5) & (df1.price_per_sqft < 12184.5)]
non_outliers
```

```python
# Box plot
sns.boxplot(data = non_outliers)
plt.title("Box plot for all the numerical columns")
plt.show()
```

# Q4. Normal Distribution #

```python
# Hist Plot
sns.histplot(data = non_outliers, x = non_outliers.price_per_sqft,bins=30, kde=True)
plt.title("Check the normality of the price per sqft column")
plt.show()
```

# Q5. Z Score Method #

```python
m = df1.price_per_sqft.mean()
s = df1.price_per_sqft.std()
# Z Score = (x - m)/s
# Z score with outliers
df1['zscore'] = (df1.price_per_sqft - m) / s
df1
```

```python
# Remove outliers using Z-scores
zscore_non_outliers = df1[(df1['zscore'] >= -3) & (df1['zscore'] <= 3)]
zscore_non_outliers
```

```python
# Droping string data columns to find proper relationship
df2 = non_outliers.drop(["location","size"],axis=1)
df2
```

```python
# Relationship
correlation=df2.corr()
correlation
```

```python
# Visualize relationships using a Heatmap
sns.heatmap(correlation, annot = True, cmap = "viridis", linewidths=1)
plt.title("Heat map for all numerical quantity")
plt.show()
```

```python
#Scatter plot between the variables to check the correlation between them
sns.pairplot(non_outliers)
plt.title("scatter plot for all columns")
plt.show()
```
